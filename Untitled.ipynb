```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
```

```python
# Sample dataset with mixed topics
documents = [
    "The cat sits on the warm mat near the window.",
    "Dogs and cats are among the most popular household pets.",
    "The lion is known as the king of the jungle.",
    "Elephants are intelligent animals with strong social bonds.",
    "I love machine learning and artificial intelligence research.",
    "Deep learning models require a lot of labeled data to train.",
    "Transformers are the foundation of modern NLP architectures.",
    "Quantum computing has the potential to revolutionize data processing.",
    "Natural Language Processing helps computers understand human language.",
    "Science explains the laws that govern the universe.",
    "Astronomy helps us understand stars, galaxies, and black holes.",
    "Space exploration leads to new discoveries and technologies.",
    "I enjoy playing football and watching live matches on weekends.",
    "Football fans cheered as their favorite team scored the winning goal.",
    "Basketball players need agility and teamwork to win games.",
    "Movies with strong character development are always enjoyable.",
    "The new superhero movie broke all box office records.",
    "Renewable energy sources help reduce environmental pollution."
]
```

```python
# Step 1: Convert text into TF-IDF features
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(documents)
```

```python
# Step 2: Dimensionality Reduction using Truncated SVD (LSA)
svd = TruncatedSVD(n_components=2, random_state=42)
X_svd = svd.fit_transform(X)
```

```python
# Step 3: Assign manual topic groups for visualization
topics = [
    "Animals", "Animals", "Animals", "Animals", 
    "Tech", "Tech", "Tech", "Tech", "Tech",
    "Science", "Science", "Science",
    "Sports", "Sports", "Sports",
    "Movies", "Movies", "Science"
]
```

```python
# Step 4: Create a professional scatter plot
plt.figure(figsize=(10, 7))
sns.set(style="whitegrid", font_scale=1.2)

palette = {
    "Animals": "#FFB74D",
    "Tech": "#64B5F6",
    "Science": "#81C784",
    "Sports": "#E57373",
    "Movies": "#BA68C8"
}
# Plot each topic group
for topic in set(topics):
    idx = [i for i, t in enumerate(topics) if t == topic]
    plt.scatter(X_svd[idx, 0], X_svd[idx, 1],
                label=topic,
                s=100, alpha=0.8, edgecolor='black',
                color=palette[topic])

# Add annotations (document numbers)
for i, text in enumerate(documents):
    plt.text(X_svd[i, 0]+0.01, X_svd[i, 1]+0.01, f"D{i+1}", fontsize=8, color='black')

# Step 5: Final plot styling
plt.title("Dimensionality Reduction in NLP using SVD (Latent Semantic Analysis)", fontsize=15, fontweight='bold')
plt.xlabel("SVD Component 1", fontsize=12)
plt.ylabel("SVD Component 2", fontsize=12)
plt.legend(title="Topic Category", loc="best")
plt.tight_layout()
plt.show()
```
